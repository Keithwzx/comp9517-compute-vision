```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# read the img
img=cv2.imread('light_rail.jpg')
```

```python
I=cv2.imread('task1_cv.png', 0)
```

```python
# get the size of the picture
rows, cols = I.shape
rows, cols
```

```python
# try
import sys
mask = np.zeros((rows,cols), dtype='uint8')
mask[1,2]=1
mask
```

```python
# define a function. The first parameter is window_size: 3,5,7...
# The second parameter is 'median' or 'max' or ''min of the list of the most frequent pixel
def spatial_filting(window_size, num_position):
    for row in range(rows):
        for col in range(cols):
            # set a mask
            mask = np.zeros((rows,cols), dtype='uint8')
            # set mask's border, if the neighbours are not in the picture, not count
            L=list(range(1,int((window_size-1)/2)+1))
            for i in L:
                # the grip of up, down, left, right
                if row+i <= rows-1:
                    mask[row+1, col]=1
                if row-i >= 0:
                    mask[row-i, col]=1
                if col+i <= cols-1:
                    mask[row, col+i]=1
                if col-i >= 0:
                    mask[row, col-i]=1
                    
                for j in L:
                    # the grip of catercorner
                    if row+i <= rows-1 and col+j <= cols-1:
                        mask[row+i, col+j]=1
                    if row-i >= 0 and col+j <= cols-1:
                        mask[row-i, col+j]=1
                    if row+i <= rows-1 and col-j >=0:
                        mask[row+i, col-j]=1
                    if row-i >= 0 and col+j >= 0:
                        mask[row-i, col-j]=1
            hist = cv2.calcHist([I], [0], mask, [256], [0,256])
#             print(np.max(hist))
            max_pixels=np.where(hist==np.max(hist))[0]
            # use median of the list of the most frequent pixel value as the centre point's pixel
            if num_position=='median':
                J[row][col]=np.median(max_pixels)
            if num_position=='max':
                J[row][col]=np.max(max_pixels)
            if num_position=='min':
                J[row][col]=np.min(max_pixels)
```

```python
# create a copy of I
J=I.copy()
spatial_filting(3, 'median')
plt.imshow(J, 'gray')
cv2.imwrite('task2_cv_median_3x3.png', J)
```

```python
# create a copy of I
J=I.copy()
spatial_filting(3, 'max')
plt.imshow(J, 'gray')
cv2.imwrite('task2_cv_max_3x3.png', J)
```

```python
# create a copy of I
J=I.copy()
spatial_filting(3, 'min')
plt.imshow(J, 'gray')
cv2.imwrite('task2_cv_min_3x3.png', J)
```
